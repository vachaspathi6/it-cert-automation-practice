```python
import plotly.express as px
import pandas as pd
import numpy as np

# Load the Gapminder dataset
df = px.data.gapminder()

# Task i: Clean the data by removing/dropping rows where lifeExp is missing
df = df.dropna(subset=['lifeExp'])

# Task ii: Create box plots and violin plot for lifeExp by continent
fig_box = px.box(df, x='continent', y='lifeExp', points='all', title='Box Plot for lifeExp by Continent')
fig_violin = px.violin(df, x='continent', y='lifeExp', box=True, title='Violin Plot for lifeExp by Continent')

# Task iii: Visualize an interactive bar plot for df
fig_bar = px.bar(df, x='continent', y='lifeExp', title='Mean lifeExp by Continent')

# Task iv: Plot a scatter chart for lifeExp and gdpPercap, colored by continent, and normalized gdpPercap
df['norm_gdpPercap'] = df['gdpPercap'] / df['gdpPercap'].max()
fig_scatter = px.scatter(df, x='lifeExp', y='gdpPercap', color='continent', size='norm_gdpPercap',
                         hover_data=['country'], title='Scatter Plot: lifeExp vs. gdpPercap')

# Task v: Create a bubble chart on lifeExp and gdpPercap, with bubble size based on population and colored by country
fig_bubble = px.scatter(df, x='lifeExp', y='gdpPercap', size='pop', color='country', hover_data=['continent'],
                        title='Bubble Chart: lifeExp vs. gdpPercap (Bubble Size = Population)')

# Task vi: Create a timeline animation for lifeExp and gdpPercap for every change
fig_timeline = px.scatter(df, x='lifeExp', y='gdpPercap', size='pop', color='country', hover_data=['continent'],
                          animation_frame='year', animation_group='country',
                          title='Timeline Animation: lifeExp and gdpPercap by Year (Bubble Size = Population)')

# Task vii: Create a 3D plot on attributes lifeExp, pop, and gdpPercap
fig_3d = px.scatter_3d(df, x='lifeExp', y='pop', z='gdpPercap', color='continent',
                       title='3D Scatter Plot: lifeExp vs. pop vs. gdpPercap')

# Show the plots
fig_box.show()
fig_violin.show()
fig_bar.show()
fig_scatter.show()
fig_bubble.show()
fig_timeline.show()
fig_3d.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Pokemon.csv')

# Task 1: Attack and Defense Powers
def task1():
    plt.figure(figsize=(10, 6))
    plt.scatter(df['Attack'], df['Defense'], alpha=0.5, color='purple')  # Scatter plot with purple color
    plt.xlabel('Attack')
    plt.ylabel('Defense')
    plt.title('Attack and Defense Powers Analysis')
    plt.grid(True)
    plt.show()

# Task 2: Pokémon Types and Their Attacks
def task2():
    plt.figure(figsize=(12, 6))
    df_type_counts = df['Type 1'].value_counts()
    sns.barplot(x=df_type_counts.index, y=df_type_counts.values, palette='muted')  # Use muted color palette for bar plot
    plt.xlabel('Pokémon Type')
    plt.ylabel('Count')
    plt.title('Pokémon Types and Their Attacks')
    plt.xticks(rotation=45)
    plt.show()

# Task 3: Attack Distribution
def task3():
    plt.figure(figsize=(8, 6))
    plt.hist(df['Attack'], bins=20, edgecolor='black', color='teal')  # Histogram with teal color
    plt.xlabel('Attack')
    plt.ylabel('Frequency')
    plt.title('Attack Distribution')
    plt.grid(True)
    plt.show()

# Task 4: Solidity Relationship between Attack and Defense
def task4():
    plt.figure(figsize=(10, 6))
    plt.hexbin(df['Attack'], df['Defense'], gridsize=20, cmap='viridis')  # Hexbin plot with viridis color map
    plt.xlabel('Attack')
    plt.ylabel('Defense')
    plt.title('Solidity Relationship between Attack and Defense')
    plt.colorbar(label='Density')
    plt.show()

# Task 5: Count of Different Pokémon Types
def task5():
    plt.figure(figsize=(12, 6))
    df['Type 1'].value_counts().plot(kind='bar', color=sns.color_palette('pastel'))  # Use pastel color palette for bar plot
    plt.xlabel('Pokémon Type')
    plt.ylabel('Count')
    plt.title('Count of Different Pokémon Types')
    plt.xticks(rotation=45)
    plt.show()

# Call the functions for each task
task1()
task2()
task3()
task4()
task5()
```
